```python
from PIL import Image
from IPython.display import display
import numpy as np

head = Image.open("./head.jpeg").convert("RGBA")
flag = Image.open("./flag.jpg").convert("RGBA")


print("原始图像")
display(head)
display(flag)

headx, heady = head.size
flagx, flagy = flag.size
print("{} {} {} {}".format(headx, heady, flagx, flagy))

grayFlag = flag.convert("L")
grayFlag = np.array(grayFlag)

grayFlag[grayFlag > 50] = 255


resizedHead = head.resize(flag.size, Image.ANTIALIAS)
display(resizedHead)

headLayer = resizedHead.copy()

maskLayer = Image.fromarray(np.uint8(grayFlag))

headLayer.paste(flag, (0, 0), mask= maskLayer)

display(headLayer)

```
